```python
data = []
human_text=""
assistant_text=""

for message in messages:
    if message["role"] == "user":
        human_text = f"### Human: \"{message['content']}\""
    elif message["role"] == "assistant":
        assistant_text = f"### Assistant: \"{message['content']}\""
    else:
        continue  # Skip messages with other roles

    if human_text and assistant_text:  # Ensure both roles are present
        combined_text = "\n".join([human_text, assistant_text])
        data.append({"text": combined_text})

with open("caption-generator-dataset.csv", "w", newline="") as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=["text"])
    writer.writeheader()
    writer.writerows(data)

print("CSV file created successfully!")
```
